```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
bujji=pd.read_csv('energy.csv')
print(bujji)
#line chart
#Andhra,Karnataka
srh=bujji[bujji['State']=='Andhra']
rcb=bujji[bujji['State']=='Karnataka']
fig,diag=plt.subplots(2)
#date,consumption
diag[0].plot(srh['Date'],srh['Consumption'],color='yellow')
diag[1].plot(rcb['Date'],rcb['Consumption'],color='black')
#bar chart
#Telangana,Tripura
srh=bujji[bujji['State']=='Telangana']
rcb=bujji[bujji['State']=='Tripura']
#date,consumption
fig,diag=plt.subplots(2)
diag[0].bar(srh['Date'],srh['Consumption'],color='yellow')
diag[1].bar(rcb['Date'],rcb['Consumption'],color='black')
#pie chart
#Kerala,Delhi
cse4='Kerala'
cse5='Delhi'
tdp=[cse4,cse5]

srh=bujji[bujji['State']=='Krala']['Consumption'].sum()
rcb=bujji[bujji['State']=='Delhi']['Consumption'].sum()
ysr=[srh,rcb]

fig,diag=plt.subplot(2)
diag[0].pie(ysr,labels=tdp)
```
